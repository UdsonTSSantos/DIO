```python
print('==== Aprendendo sobre listas =======')
print('Listas são usadas entre cochetes')
print('Vamos criar nossa primeira lista chamada animais'.upper())
animais = [1, 2, 3]
print('\n Inserindo novos valores na lista animais'.upper())
animais = ['cachorro', 'gato', 12345, 6.5]
print(animais)
print('\n Agora vamos imprimir somente o 1 elemento da nossa lista animais, no caso tem de sair cachorro.')
print(' O elemento zero é: ',animais[0], '| O zero entre cochetes indica a posição do elemento que queremos.')
per = (input('\n Deseja ver como se faz para imprimir o 1 elemento da lista, [s/n]'))
if per == 's':
   print(' Ok, vamos imprimir o 1 elemento da lista animais.')
   print('\n O 1 elemento da lista animais é: ', animais[1], '| até aqui tudo sobre controle, correto? ')
else:
    print('Tudo bem, se já entendeu o conteúdo ótimo, vamos para a próxima etapa, substituição de elementos em listas.')

print('Vamos aprender como substituir elementos de uma lista.'.upper())
print('No nosso exemplo vamos substituir o GATO que é o elemento 1 por girafa que não tem na nossa lista.')
per1 = input('Deseja continuar e aprender como fazer a substituição? [s/n]')
if per1 == 's':
  print('Vamos lá, na nossa lista temos 4 elementos, sendo eles com posições de 0,1,2,3')
  print('O gato é o elemento de número 1 porque em Python se começa do zero a contar os elementos.')
  print('O código fica assim:  animais[1] = girafa ')
  print('Explicando, animais(nome da variável de tem os elementos) | [1] identificação do elemento na lista que vamos substituir')
  animais[1] = 'girafa'
  print(animais)
else:
  print('Vamos para a próxima etapa, como remover um elemento de uma lista usando o comando .remove()')
  print('O código fica assim: animais.remove(cachorro)')
  animais.remove('cachorro')
  print(animais)
```
